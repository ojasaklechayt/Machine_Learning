```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 Countries that uses zomato
plt.pie(country_values[:3], labels=country_names[:3], autopct="%1.2f%%")
```

Observation: Zomato maximum records or transactions are from India
After that USA and then United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When the rating is between 4.5 to 4.9 -> Excellent
2. When the rating is between 4.0 to 4.4 -> Very Good
3. When the rating is between 3.5 to 3.9 -> Good
4. When the rating is between 2.5 to 3.4 -> Average
5. When the rating is between 1.8 to 2.4 -> Poor
6. When the rating is 0 -> Not Given

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color', data=ratings, palette=['gray','red','orange','yellow','green','green'])
```

### Observations:
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x='Rating color', data=ratings,hue='Rating color', palette=['gray','red','orange','yellow','green','green'])
```

```python
### Find the countries name that has given 0 rating
final_df[final_df['Rating color']== 'White'].groupby('Country').size().reset_index()
```

## Obvservations: 
Maximum number of 0 ratings are from Indian customers
